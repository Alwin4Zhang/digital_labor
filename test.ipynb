```python
# 存储桶和文件配置
bucket_name = "digital-labor"  # 替换为你的存储桶名称
file_path = "/Users/bello/Downloads/测试图片/2298172.jpg"  # 本地图片路径
object_name = "2298172.jpg"  # 上传后的对象名称
content_type = "image/jpeg"  # 根据图片类型调整（如image/png）
```

```python
if minio_client.bucket_exists(bucket_name=bucket_name):  # bucket_exists：检查桶是否存在
    print("该存储桶已经存在")
else:
    minio_client.make_bucket(bucket_name=bucket_name)
    print("存储桶创建成功")
```

```python
# 上传图片到Minio
minio_client.fput_object(
    bucket_name=bucket_name,
    object_name=object_name,
    file_path=file_path,
    content_type=content_type
)

print(f"图片 {file_path} 上传成功")
```

```python
# 生成可直接预览的URL（预签名，有效期7天）
preview_url = minio_client.get_presigned_url(
    "GET",
    bucket_name,
    object_name
)
print("可预览的URL：", preview_url)
```
